```python
# For loops
```

```python
numbers = [2,3,4,5]
sum = 0
```

```python
for i in numbers:
    sum = sum + i
```

```python
print("The sum of numbers vector is:",  sum)
```

```python
# Print text:
text = "Running"
for j in text:
    print(j)
```

```python
# while loop
counter = 1
while counter < 5:
    print(counter)
    counter = counter + 1
```

```python
# same outcome as above, except via for loop (using range):
for dog in range(1,5):
    print(dog)
```

```python
# Create multiplication table:
num = int(input("Enter an integer:"))
for index in range(1,11):
    product = index * num
    print(num, "x", index, "=", product)
```

```python
# homework: find sum of numbers 1 to 100:
cumsum = 0
for index in range(1,101):
    cumsum = cumsum + index
print(cumsum)
```

```python
# generalize range
cumsum = 0
for index in range(1,6,2):
    cumsum = cumsum + index
print(cumsum)
```

```python
# print
dog = list(range(10))
print(dog)
```

```python
# indexing
fish = ["cod", "salmon", "walleye"]
for i in range(len(fish)):
    print("I like", fish[i])
```

```python
# find my favorite food
eaters_name = "Michael B"
foods = {"doggies" : "dog food", "kitties" : "cat food", "Michael B" : "steak",
        "Norman" : "green eggs and ham"}
for counter in foods:
    if counter == eaters_name:
        print(foods[counter])
        break
else:
    print("I can't find that person's favorite food")
```
